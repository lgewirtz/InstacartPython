## 4. Create 'region' column based on 'State' column; determine if difference in spending

```python
# drop _merge column 

dfm = dfm.iloc[: , :-1]
```

```python
dfm.head()
```

```python
# create list 'result2' with region corresponding to 'State' for full dataset 'dfm'

result2 = []

for value in dfm["State"]:
  if (value == 'Alabama') | (value == 'Delaware') | (value == 'Louisiana') | (value == 'Arkansas') | (value == 'Louisiana')  | (value == 'District of Columbia') | (value == 'Delaware')  | (value == 'Florida') | (value == 'Georgia') | (value == 'Kentucky') | (value == 'Louisiana') | (value == 'Maryland')  | (value == 'Mississippi') | (value == 'North Carolina') | (value == 'Oklahoma') | (value == 'South Carolina')  | (value == 'Tennessee') | (value == 'Texas') | (value == 'Virginia') | (value == 'West Virginia'):
    result2.append("South")
  elif (value == 'Iowa') | (value == 'Illinois') | (value == 'Indiana') |  (value == 'Kansas') | (value == 'Michigan') | (value == 'Minnesota') | (value == 'Missouri') | (value == 'North Dakota') | (value == 'Nebraska') | (value == 'Ohio') | (value == 'South Dakota') | (value == 'Wisconsin'):
    result2.append("Midwest")
  elif (value == 'Connecticut') | (value == 'Massachusetts') | (value == 'Maine') |  (value == 'New Hampshire') | (value == 'New Jersey') | (value == 'New York') | (value == 'Pennsylvania') | (value == 'Rhode Island') | (value == 'Vermont'):
    result2.append("Northeast")
  elif (value == 'Alaska') | (value == 'Arizona') | (value == 'California') |  (value == 'Colorado') | (value == 'Hawaii') | (value == 'Idaho') | (value == 'Montana') | (value == 'New Mexico') | (value == 'Nevada') | (value == 'Oregon') | (value == 'Utah') | (value == 'Washington') | (value == 'Wyoming'):
    result2.append("West")
  else:
    result2.append("Where do I live?")
```

```python
# create column 'region' in full dataset, dfm

dfm['region'] = result2
```

```python
#Save dataset with 'region' column added

dfm.to_pickle(os.path.join(path, '02 Data', 'Prepped Data', 'dfmregionadded.pkl'))
```

```python
dfm.head()
```

```python
dfm['region'].value_counts(dropna = False)
```

```python
# create df to compare 'region' & 'spending_flag' to if there is a correlation

crosstab = pd.crosstab(dfm['region'], dfm['spending_flag'], dropna = False)
```

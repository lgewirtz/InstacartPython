# 3. Create loyalty flag

```python
# Group by department, find max of order_number for each group

dfm['max_order'] = dfm.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
dfm.head(15)
```

There were no pink warning signs because the script is writing to the whole DF and not a slice

```python
pd.options.display.max_rows = None
```

```python
dfm.head(100)
```

```python
dfm.loc[dfm['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
dfm.loc[(dfm['max_order'] <= 40) & (dfm['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
dfm.loc[dfm['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
dfm.head()
```

```python
dfm['loyalty_flag'].value_counts(dropna = False)
```

In the full data set there is a smaller proportion of New customers compared to the data subset

```python
dfm[['user_id', 'order_number', 'loyalty_flag', 'max_order' ]].head(30)
```

```python
# Find basic price statistics per customer grouped by loyalty_flag

dfm.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

loyal customers buy products with a lower average price

# 4. Create spending flag

```python
# Group by user_id, find mean of price for each group

dfm['user_price_mean'] = dfm.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
dfm.head()
```

Break users into groups of 'Low spender' or 'High spender'

```python
dfm.loc[dfm['user_price_mean'] < 10, 'spending_flag'] = 'Low spender'
```

```python
dfm.loc[(dfm['user_price_mean'] >= 10), 'spending_flag'] = 'High spender'
```

```python
dfm.head(20)
```

```python
dfm['spending_flag'].value_counts(dropna=False)
```

```python
print(dfm[['spending_flag', 'user_price_mean', 'prices']].head(100))
```

# 5. Create order_frequency flag

```python
# Find median of "days_since_prior_order" grouped by 'user_id'

dfm['med_days_since'] = dfm.groupby(['user_id'])['days_since_prior_order'].transform(np.mean)
```

```python
print(dfm[['med_days_since', 'user_id', 'days_since_prior_order']].head(30))
```

```python
# Create 'Order frequency' flag
```

```python
dfm.loc[dfm['med_days_since'] > 20, 'Order freq'] = 'Non-frequent customer'
```

```python
dfm.loc[(dfm['med_days_since'] > 10) & (dfm['med_days_since'] <= 20), 'Order freq'] = 'Regular customer'
```

```python
dfm.loc[dfm['med_days_since'] <= 10, 'Order freq'] = 'Frequent customer'
```

```python
dfm.head()
```

```python
print(dfm[['med_days_since', 'user_id', 'Order freq']].head(30))
```

```python
dfm['Order freq'].value_counts(dropna = False)
```
